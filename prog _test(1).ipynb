```python
# Faire une fonction qui traite les dataframe
def traitement(df):
    df['Dépenses'] = df['Dépenses'].str.strip()
    df['Dépenses'] = df['Dépenses'].str.replace(',', '.')

    df['Solde'] = df['Solde'].str.strip()
    df['Solde'] = df['Solde'].str.replace(',', '.')

    # supprimer -+ dans la colonne Solde
    #transfert14['Solde'] = transfert14['Solde'].str.replace('+-0', '0')

    df['Revenus'] = df['Revenus'].str.strip()
    df['Revenus'] = df['Revenus'].str.replace(',', '.')

    for i in range(len(df)):
        if 'mio. €' in df['Dépenses'][i]:
            df['Dépenses'][i] = df['Dépenses'][i].replace(' mio. €', '')
            df['Dépenses'][i] = float(df['Dépenses'][i])
            df['Dépenses'][i] = df['Dépenses'][i] * 1000000
        elif 'K €' in df['Dépenses'][i]:
            df['Dépenses'][i] = df['Dépenses'][i].replace(' K €', '')
            df['Dépenses'][i] = float(df['Dépenses'][i])
            df['Dépenses'][i] = df['Dépenses'][i] * 1000
        elif 'mrd. €' in df['Dépenses'][i]:
            df['Dépenses'][i] = df['Dépenses'][i].replace(' mrd. €', '')
            df['Dépenses'][i] = float(df['Dépenses'][i])
            df['Dépenses'][i] = df['Dépenses'][i] * 1000000000
        elif ' €' in df['Dépenses'][i]:
            df['Dépenses'][i] = df['Dépenses'][i].replace(' €', '')
            df['Dépenses'][i] = float(df['Dépenses'][i])
        elif '-' == df['Dépenses'][i]:
            df['Dépenses'][i] = df['Dépenses'][i].replace('-', '')
            df['Dépenses'][i] = 0.0

        if 'mio. €' in df['Solde'][i]:
            df['Solde'][i] = df['Solde'][i].replace(' mio. €', '')
            df['Solde'][i] = float(df['Solde'][i])
            df['Solde'][i] = df['Solde'][i] * 1000000
        elif 'K €' in df['Solde'][i]:
            df['Solde'][i] = df['Solde'][i].replace(' K €', '')
            df['Solde'][i] = float(df['Solde'][i])
            df['Solde'][i] = df['Solde'][i] * 1000
        elif 'mrd. €' in df['Solde'][i]:
            df['Solde'][i] = df['Solde'][i].replace(' mrd. €', '')
            df['Solde'][i] = float(df['Solde'][i])
            df['Solde'][i] = df['Solde'][i] * 1000000000
        elif ' €' in df['Solde'][i]:
            df['Solde'][i] = df['Solde'][i].replace(' €', '')
            df['Solde'][i] = float(df['Solde'][i])
        elif '+-' in df['Solde'][i]:
            df['Solde'][i] = df['Solde'][i].replace('+-', '')
            df['Solde'][i] = float(df['Solde'][i])


        if 'mio. €' in df['Revenus'][i]:
            df['Revenus'][i] = df['Revenus'][i].replace(' mio. €', '')
            df['Revenus'][i] = float(df['Revenus'][i])
            df['Revenus'][i] = df['Revenus'][i] * 1000000
        elif 'K €' in df['Revenus'][i]:
            df['Revenus'][i] = df['Revenus'][i].replace(' K €', '')
            df['Revenus'][i] = float(df['Revenus'][i])
            df['Revenus'][i] = df['Revenus'][i] * 1000
        elif 'mrd. €' in df['Revenus'][i]:
            df['Revenus'][i] = df['Revenus'][i].replace(' mrd. €', '')
            df['Revenus'][i] = float(df['Revenus'][i])
            df['Revenus'][i] = df['Revenus'][i] * 1000000000
        elif ' €' in df['Revenus'][i]:
            df['Revenus'][i] = df['Revenus'][i].replace(' €', '')
            df['Revenus'][i] = float(df['Revenus'][i])
        
        elif '-' == df['Revenus'][i]:
            df['Revenus'][i] = df['Revenus'][i].replace('-', '')
            df['Revenus'][i] = 0.0
    
```
